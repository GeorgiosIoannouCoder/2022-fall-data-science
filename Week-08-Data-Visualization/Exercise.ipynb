# Using the `Sales.csv` data:

* Make a time series chart that shows the sales for each item over the past 12 months.
* What does this tell you about the items?
* Which item would you recommend they push more of and which item/s would you recommend they stop selling?
* Show what percentage of each item contributes to total sales.


```python
# Load the Sales.csv data into a dataframe.
# Read the file Sales.csv located inside the data folder and then load the data.
sales_df = pd.read_csv('data/Sales.csv')

# Print/Display/Return the rows of the file Sales.csv to make sure the file was loaded successfully.

sales_df
```

```python
# Print the shape.

sales_df.shape
```

```python
# Inspect nulls and duplicates.

print('NULLS')
print(sales_df.isnull().sum(), '\n')
print('Duplicates')
print('sales_df.duplicated().sum() =', sales_df.duplicated().sum())
```

```python
# Make a time series chart that shows the sales for each item over the past 12 months.
# Exlude the month_number column from the plot.

cols = list(sales_df.columns)
cols.remove('month_number')

fig, ax = plt.subplots(figsize = (19, 10))
sales_df.plot(kind="line", x = "month_number", y = cols, ax = ax)

ax.set_xlabel('Month')
ax.set_ylabel('Sales')
ax.set_title('Sales per Month')
plt.legend()
plt.show()
```

___
<h2 align="center"> <span style="color:#964B00"> Bathsoap is being sold the most among the rest of the products. <br> Bathsoap is being sold the most during the end of the year. (October-December) </span> </h2>

___


___
<h2 align="center"> <span style="color:#964B00"> I would recommend to push bathsoap more and stp selling face cream. </span> </h2>

___


```python
# Show what percentage of each item contributes to total sales.

percentage_list = []
total_sales = 0

for item in cols:
    total_sales += sales_df[item].sum()

for item in cols:
    percentage_list.append((sales_df[item].sum() / total_sales) * 100)
    
percentage_array = np.array(percentage_list)

fig, ax = plt.subplots(figsize = (19, 10))
plt.pie(percentage_array, labels = cols, autopct = '%.2f%%')
plt.legend(loc='upper left')
plt.show()
```
