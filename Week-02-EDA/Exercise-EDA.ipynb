## It seems as though in the years before (and includig) 2000, they were counting in tens.

Make a new column that is called `adjusted_price`, that contains all prices, however, for every year before 2000 make it 10x the original MSRP.

- Hint; you might need to use our old friend `np.where`


```python
# Make a column where is 10 when year is less than 2000, else 1.
df['price_adjuster'] = np.where(df['Year'] <= 2000, 10, 1)

# Multiply the MSRP by the price adjuster.
df['adjusted_price'] = df['MSRP'] * df['price_adjuster']

df.head() # Print/Display/Return the first 5 rows of the file cars.csv with the new column adjusted_price.
```

## Replot the new adjusted price.

- Make the y-axis start at 0 and go up to 100,000


```python
# Plot new prices.
prices = df.groupby('Year')['adjusted_price'].mean()
prices.plot(ylim=(0, 100000), figsize=(17, 7), title='Average MSRP vs. Year');
```

## Remove any outliers from Engine HP and highway MPG

<img src='https://miro.medium.com/max/1400/1*2c21SkzJMf3frPXPAR_gZA.png' width=500>

- Outliers meaning values that are outside 1.5x the Inter Quartile Range (see image above).
- For each column (Engine HP and highway MPG):
- Calculate the 0.25 and 0.75 Quantiles
- Calculate the Inter Quartile Range (IQR)
- Create condition mask for the values that are outliers below (in the 'Minimum' range).
- Create condition mask for the values that are outliers above (in the 'Maximum' range).
- Filter the dataframe to remove any values that are in the above section _OR_ the below section. (hint; it may be easier to use the inverse selection '~'.
- Make the same boxplots of Engine HP and Highway MPG as before but with the this dataframe.


```python
# Remove any  outliers from Engine HP and highway MPG.
engine_hp_25 = np.percentile(temp_df['Engine HP'], 25) # 25th Percentile of the Engine HP.
engine_hp_75 = np.percentile(temp_df['Engine HP'], 75) # 75th Percentile of the Engine HP.

engine_hp_iqr = engine_hp_75 - engine_hp_25 # Inter Quartile Range of the Engine HP.

highway_mpg_25 = np.percentile(temp_df['highway MPG'], 25) # 25th Percentile of the highway MPG.
highway_mpg_75 = np.percentile(temp_df['highway MPG'], 75) # 75th Percentile of the highway MPG.

highway_mpg_iqr = highway_mpg_75 - highway_mpg_25 # Inter Quartile Range of the highway MPG.

# Remove any  outliers from Engine HP and highway MPG.
temp_df = temp_df[(temp_df['Engine HP'] < (engine_hp_75 + (engine_hp_iqr * 1.5)))
                  & (temp_df['Engine HP'] > (engine_hp_25 - (engine_hp_iqr * 1.5)))]
temp_df = temp_df[(temp_df['highway MPG'] < (highway_mpg_75 + (highway_mpg_iqr * 1.5)))
                  & (temp_df['highway MPG'] > (highway_mpg_25 - (highway_mpg_iqr * 1.5)))]
```

- In the column names, replace all the spaces with an underscore, and make them all lowercase as well.


```python
# * In the column names, replace all the spaces with an underscore, and make them all lowercase as well.
new_df_column_names_list = []         # new_df_column_names_list stores the new column names.

for column in df.columns:
    # For each column name in the original dataframe, replace all of its spaces with underscores and make all letters lowercase.
    # Finally, append the new column name in the new_df_column_names_list.
    new_df_column_names_list.append(column.replace(' ', '_').lower())

df.columns = new_df_column_names_list # Replace the old column names with the new column names.

df.columns                            # Display the column names of the original dataframe.
```
